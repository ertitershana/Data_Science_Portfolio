## 4. Analyse losing price change in apple stock


##### Daily Stock Return Formula 
To determine your daily gain or loss for a stock, subtract the opening price from the closing price. Then, multiply the result by the number of shares you own in the company.

```python
company_list
```

## 7. Analysing whether Daily change in Closing price of stocks or Daily Returns in Stock are co-related

```python
closing_price
```

```python
closing_price['apple_close']
```

```python
closing_price['apple_close'].shift(1)
```

```python
(closing_price['apple_close'] - closing_price['apple_close'].shift(1))/closing_price['apple_close'].shift(1) * 100
```

```python
for col in closing_price.columns:
    closing_price[col + '_pct_change'] = (closing_price[col] - closing_price[col].shift(1))/closing_price[col].shift(1) * 100
```

```python
closing_price
```

```python
closing_price.columns
```

```python
clsing_p = closing_price[['apple_close_pct_change', 'amzn_close_pct_change',
       'goog_close_pct_change', 'msft_close_pct_change']]
```

```python
clsing_p
```

```python
g = sns.PairGrid(data= clsing_p)
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```

A linear relationship is observed between 'AAPL_close_pct_change' to 'AMZN_close_pct_change'.
